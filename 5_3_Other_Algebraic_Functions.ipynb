**Definition 5.4.** Let $x$ be a real number and $n$ a natural number. If $n$ is odd, the **principal $\boldsymbol{n^{th}}$ root** of $x$, denoted $\sqrt[n]{x}$ is the unique real number satisfying $(\sqrt[n]{x})^n=x$. If $n$ is even, $\sqrt[n]{x}$ is defined similarly provided $x\geq 0$ and $\sqrt[n]{x}\geq 0$. The **index** is the number $n$ and the **radicand** is the number $x$. For $n=2$, we write $\sqrt{x}$ instead of $\sqrt[2]{x}$.

**Theorem 5.6. Properties of Radicals:** Let $x$ and $y$ be real numbers and $m$ and $n$ be natural numbers. If $\sqrt[n]{x}$, $\sqrt[n]{y}$ are real numbers, then

- **Product Rule:** $\sqrt[n]{xy}=\sqrt[n]{x}\sqrt[n]{y}$

- **Powers of Radicals:** $\sqrt[n]{x^m}=(\sqrt[n]{x})^m$

- **Quotient Rule:** $\sqrt[n]{\frac{x}{y}}=\frac{\sqrt[n]{x}}{\sqrt[n]{y}}$, provided $y\ne 0$.

- If $n$ is odd, $\sqrt[n]{x^n}=x$; if $n$ is even, $\sqrt[n]{x}=|x|$.

**Definition 5.5.** Let $x$ be a real number, $m$ an integer and $n$ a natural number.

- $x^{\frac{1}{n}}=\sqrt[n]{x}$ and is defined whenever $\sqrt[n]{x}$ is defined.

- $x^{\frac{m}{n}}=(\sqrt[n]{x}^m)=\sqrt[n]{x^m}$, whenever $(\sqrt[n]{x})^m$ is defined.

### 1. $f(x)=\sqrt{1-x^2}$

```python
x = np.linspace(-1, 1, 10000)
f = np.sqrt(1 - x**2)

fig, axs = plt.subplots()

axs.plot(x, f, label=r'$f(x)=\sqrt{1-x^2}$')

axs.axis('equal')
axs.set(xlim=(-2,2), ylim=(-2,2))

axs.axhline(0, color='black', linewidth=0.25)
axs.axvline(0, color='black', linewidth=0.25)

axs.legend()
plt.show()
```

### 2. $f(x)=\sqrt{x^2-1}$

```python
x = np.linspace(-4, 4, 10000)
f = np.sqrt(x**2 - 1)

fig, axs = plt.subplots()

axs.plot(x, f, label=r'$f(x)=\sqrt{x^2-1}$')

axs.set_aspect('equal', 'box')

axs.axhline(0, color='black', linewidth=0.25)
axs.axvline(0, color='black', linewidth=0.25)

axs.legend()
plt.show()
```

### 3. $f(x)=x\sqrt{1-x^2}$

```python
x = np.linspace(-2, 2, 10000)
f = x * np.sqrt(1 - x**2)

fig, axs = plt.subplots()

axs.plot(x, f, label=r'$f(x)=x\sqrt{1-x^2}$')

axs.set_aspect('equal', 'box')

axs.axhline(0, color='black', linewidth=0.25)
axs.axvline(0, color='black', linewidth=0.25)

axs.legend()
plt.show()
```

### 4. $f(x)=x\sqrt{x^2-1}$

```python
x = np.linspace(-2, 2, 10000)
f = x * np.sqrt(x**2 - 1)

fig, axs = plt.subplots()

axs.plot(x, f, label=r'$f(x)=x\sqrt{x^2-1}$')

axs.axis('equal')

axs.axhline(0, color='black', linewidth=0.25)
axs.axvline(0, color='black', linewidth=0.25)

axs.legend()
plt.show()
```

### 5. $f(x)=\sqrt[4]{\frac{16x}{x^2-9}}$

```python
x = np.linspace(-10, 10, 10000)
f = np.sqrt(16*x / (x**2-9))

fig, axs = plt.subplots()

axs.plot(x, f, label=r'$f(x)=\sqrt[4]{\frac{16x}{x^2-9}}$')

axs.axis('equal')
axs.set(xlim=(-10,10), ylim=(-10,10))

axs.axvline(-3, color='orange', linestyle='dashed', label=r'x=-3')
axs.axvline(3, color='orange', linestyle='dashed', label=r'x=3')

axs.axhline(0, color='black', linewidth=0.25)
axs.axvline(0, color='black', linewidth=0.25)

axs.legend()
plt.show()
```
