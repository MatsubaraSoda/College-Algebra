**Definition 1.4.** A **relation** is a set of points in the plane.

---

<center><b>Equations of Vertical and Horizontal Lines</b></center>

- The graph of the equation $x=a$ is a **vertical line** through $(a,0)$.
- The graph of the equation $y=b$ is a **horizontal line** through $(0,b)$.

---

<center><b>The Fundamental Graphing Principle</b></center>

The graph of an equation is the set of points which satisfy the equation. That is, a point $(x,y)$ is on the graph of an equation if and only if $x$ and $y$ satisfy the eqution.

---

**Definition 1.5.** Suppose the graph of an equation is given.

- A point on a graph which is also on the x-axis is called an **x-intercept** of the graph.
- A point on a graph which is also on the y-axis is called an **y-intercept** of the graph.

---

<center><b>Finding the Intercepts of the Graph of an Equation</b></center>

- x-intercepts have the form $(x,0)$; set $y=0$ in the equation and solve for $x$.
- y-intercepts have the form $(0,y)$; set $x=0$ in the equation and solve for $y$.

---

<center><b>Testing the Graph of an Equation for Symmetry</b></center>

To test the graph of an equation for symmetry

- about the y-axis : substitute $(-x,y)$ into the equation and simplify. If the result is equivalent to the original equation, the graph is symmetric about the y-axis.
- about the x-axis : substitute $(x,-y)$ into the equation and simplify. If the result is equivalent to the original equation, the graph is symmetric about the x-axis.
- about the origin : substitute $(-x,-y)$ into the equation and simplify. If the result is equivalent to the original equation, the graph is symmetric about the origin.

## Example 1.2.1. Graph the following relations.

1\. $A=\{ (0,0), (-3,1), (4,2), (-3,2) \}$

2\. $HLS_1 = \{ (x,3) \mid -2 \leq x \leq 4 \}$

3\. $HLS_2 = \{ (x,3) \mid -2 \leq x \lt 4 \}$

4\. $V = \{ (3,y) \mid y \text{ is a real number} \}$

5\. $H = \{ (x,y) \mid y = -2 \}$

6\. $R = \{ (x,y) \mid 1 \lt y \leq 3 \}$

```python
# @title
import matplotlib.pyplot as plt
import numpy as np

# Question 1

# Create the set
set_A = [(0,0), (-3,1), (4,2), (-3,2)]

# Extract x and y coordinates for plotting
x_coords, y_coords = zip(*set_A)

# Plot the points
plt.scatter(x_coords, y_coords, s=20, color='blue')

# Set axis limits
plt.xticks(np.arange(-10, 11, 2))
plt.yticks(np.arange(-10, 11, 2))

# Add labels and title
plt.xlabel('x')
plt.ylabel('y')
plt.title(r'1. Graph of the Set $A=\{ (0,0), (-3,1), (4,2), (-3,2) \}$')

# Add gridlines for better visibility
plt.grid(True, linestyle='--', alpha=0.6)

# Show the cross axis and the plot
plt.axhline(0, color='black', linewidth=0.5)
plt.axvline(0, color='black', linewidth=0.5)
plt.show()

```

## Example 1.2.3 Graph $x^2+y^3=1$

```python
import numpy as np
import matplotlib.pyplot as plt

# 定义函数
def func(x):
    return np.cbrt(1 - x**2)

# 生成数据点
x = np.linspace(-10, 10, 1000)
y = func(x)

# Set axis limits
plt.xlim(-10, 10)
plt.xticks(np.arange(-10, 10, 1))
plt.ylim(-10, 10)
plt.yticks(np.arange(-10, 10, 1))

# 绘制图形
plt.plot(x, y, label=r'$x^2 + y^3 = 1$')
plt.axhline(0, color='black',linewidth=0.5)
plt.axvline(0, color='black',linewidth=0.5)
plt.grid(color = 'gray', linestyle = '--', linewidth = 0.5)
plt.title('Graph of $x^2 + y^3 = 1$')
plt.xlabel('x')
plt.ylabel('y')
plt.legend()
plt.show()
```

## 1.2.2 EXERCISES

53\. $x^3+y^3-3xy=0$ Folium of Descartes

54\. $x^4=x^2+y^2$ Kampyle of Eudoxus

55\. $y^2=x^3+3x^2$ Tschirnhausen cubic

56\. $(x^2+y^2)^2=x^3+y^3$ Crooked egg

```python
import numpy as np
import matplotlib.pyplot as plt

# Define the function
def equation(x, y):
    return x**3 + y**3 - 3*x*y

# Generate x and y values within the specified range
x = np.linspace(-3, 3, 1000)
y = np.linspace(-3, 3, 1000)

# Create a meshgrid from x and y
X, Y = np.meshgrid(x, y)

# Compute the values of the equation for each point in the meshgrid
Z = equation(X, Y)

# Plot the curve
plt.contour(X, Y, Z, levels=[0], colors='b')
plt.title('$x^3 + y^3 - 3xy = 0$')
plt.xlabel('x')
plt.ylabel('y')
plt.grid(True)

# Ensure the aspect ratio is equal
'''
Ensures that the plot has an equal aspect ratio in both x and y directions,
and the adjustment is done using a bounding box to maintain the correct data aspect ratio.
This is useful when you want to ensure
that the units along the x and y axes are represented equally in the plot.
'''
plt.gca().set_aspect('equal', adjustable='box')


plt.show()
```
