**Definition 4.1.** A **rational function** is a function which is the ratio of polynomial functions. Said differently, $r$ is a rational function if it is of the form
$$
r(x)=\frac{p(x)}{q(x)}
$$
where $p$ and $q$ are polynomial functions.

**Definition 4.2.** The line $x=c$ is called a **vertical asymptote** of the graph of a function $y=f(x)$ if as $x\to c^{-}$ or $x\to c^+$, either $f(x)\to \infty$ or $f(x)\to -\infty$.

**Definition 4.3.** The line $y=c$ is called a **horizontal asymptote** of the graph of a function $y=f(x)$ if as $x\to -\infty$ or as $x\to \infty$, $f(x)\to c$.

**Theorem 4.1. Location of Vertical Asymptotes and Holes:** Suppose $r$ is a rational function which can be written as $r(x)=\frac{p(x)}{q(x)}$ where $p$ and $q$ have no common zeros. Let $c$ be a real number which is not in the domain of $r$.

- If $q(c)\ne 0$, then the graph of $y=r(x)$ has a hole at $(c,\frac{p(x)}{q(x)})$.
- If $q(c)=0$, then the line $x=c$ is a vertical asymptote of the graph of $y=r(x)$.

**Theorem 4.2. Location of Horizontal Asymptotes:** Suppose $r$ is a rational function and $r(x)=\frac{p(x)}{q(x)}$, where $p$ and $q$ are polynomial functions with leading coefficients $a$ and $b$, respectively.

- If the degree of $p(x)$ is the same as the degree of $q(x)$, then $y=\frac{a}{b}$ is the horizontal asymptote of the graph of $y=r(x)$.
- If the degree of $p(x)$ is less than the degree of $q(x)$, then $y=0$ is the horizontal asymptote of the graph of $y=r(x)$.
- If the degree of $p(x)$ is greater than the degree of $q(x)$, then the graph of $y=r(x)$ has no horizontal asymptotes.

**Definition 4.4.** The line $y=mx+b$ where $m\ne 0$ is called a **slant asymptote** of the graph of a function $y=f(x)$ if as $x\to -\infty$ or as $x\to \infty$, $f(x)\to mx+b$.

**Theorem 4.3. Determination of Slant Asymptotes:** Suppose $r$ is a rational function and $r(x)=\frac{p(x)}{q(x)}$, where the degree of $p$ is *exactly* one more than the degree of $q$. Then the graph of $y=r(x)$ has the slant asymptote $y=L(x)$ where $L(x)$ is the quotient obtained by dividing $p(x)$ by $q(x)$.

````python
import sympy as sp
import numpy as np
import matplotlib.pyplot as plt
````

## Exercises 1 - 18

In Exercises 1 - 18, for the given rational function $f$:

- Find the domain of $f$.
- Identify any vertical asymptotes of the graph of $y=f(x)$.
- Identify any holes in the graph.
- Find the horizontal asymptote, if it exists.
- Find the slant asymptote, if it exists.
- Graph the function using a graphing utility and describe the behavior near the asymptotes.

````python
# https://matplotlib.org/stable/gallery/color/color_by_yvalue.html#sphx-glr-gallery-color-color-by-yvalue-py

x = sp.Symbol('x')
f = 1 / x

x_undefined_values = np.array([0])

x_values = np.linspace(-5,5,1000)
x_values = np.setdiff1d(x_values, x_undefined_values)

upper = -10
lower = 10

y_values = sp.lambdify(x, f, 'numpy')(x_values)
y_values = np.ma.masked_where( (y_values < upper) | (y_values > lower), y_values)

plt.plot(x_values, y_values, label=r'$f(x)=\frac{1}{x}$')
plt.xlim(-5,5)
plt.ylim(-5,5)
plt.axhline(0, color='black', linestyle='--', linewidth=0.25)
plt.axvline(0, color='black', linestyle='--', linewidth=0.25)
plt.title('Graph of Rational Function')
plt.legend()
plt.show()
````

### 1. $f(x)=\frac{x}{3x-6}$

````python
x = sp.Symbol('x')
f = x / (3*x - 6)

x_undefined_values = np.array([2])

x_values = np.linspace(-100,100,10000)
x_values = np.setdiff1d(x_values, x_undefined_values)

upper = -10
lower = 10

y_values = sp.lambdify(x, f, 'numpy')(x_values)
y_values = np.ma.masked_where( (y_values < upper) | (y_values > lower), y_values)

plt.plot(x_values, y_values, label=r'$f(x)=\frac{x}{3x-6}$')
plt.xlim(-6,6)
plt.ylim(-6,6)

plt.axvline(2, color='orange', linestyle='dashed', label=r'$x=2$')
plt.axhline(1/3, color='purple', linestyle='dashed', label=r'$y=\frac{1}{3}$')

plt.axhline(0, color='black', linestyle='solid', linewidth=0.25)
plt.axvline(0, color='black', linestyle='solid', linewidth=0.25)
plt.title('Graph of Rational Function')
plt.legend()
plt.show()
````

## Exercise 19

The cost $C$ in dollars to remove $p\%$ of the invasive species of Ippizuti fish from Sasquatch Pond is given by

$$
C(p)=\frac{1770p}{100-p},\ 0\leq p\lt 100
$$

(a) Find and interpret $C(25)$ and $C(95)$.

(b) What does the vertical asymptote at $x=100$ mean within the context of the problem?

(c) What percentage of the Ippizuti fish can you remove for $40000?

````python
p = sp.Symbol('p')
C = (1770*p) / (100-p)

p_values = np.linspace(0,100,10000)
p_values = p_values[p_values != 100]
y_values = sp.lambdify(p, C, 'numpy')(p_values)

print("C(25) =", sp.lambdify(p, C, 'numpy')(25))
print("C(95) =", sp.lambdify(p, C, 'numpy')(95))

solutions = sp.solve(sp.Eq(C,40000))
for s in solutions:
  print('p =', s.evalf())

plt.plot(p_values, y_values, label=r'$C(p)=\frac{1770p}{100-p}$')

plt.xlim(0,100)
plt.ylim(0,50000)

plt.axhline(0, color='black', linestyle='solid', linewidth=0.25)
plt.axvline(0, color='black', linestyle='solid', linewidth=0.25)
plt.title('Exercise 19')
plt.legend()
plt.show()
````
