**Therorem 3.8. Cauchy's Bound:** Suppose $f(x)=a_nx^n + a_{n-1}x^{n-1}+\ ...\ + a_1x+a_0$ is a polynomial of degree $n$ with $n\geq 1$. Let $M$ be the largest of the numbers: $\frac{|a_0|}{|a_n|},\frac{|a_1|}{|a_n|},\ ...\ ,\frac{|a_{n-1}|}{|a_n|}.$ Then all the real zeros of $f$ lie in in the interval $[-(M+1),M+1]$.

> The proof of this fact is not easily explained within confines of this text.

---

**Theorem 3.9. Rational Zeros Theorem:** Suppose $f(x)=a_n x^n + a_{n-1}x^{n-1}+\ ...\ + a_1x+a_0$ is a polynomial of degree $n$ with $n\geq 1$, and $a_0,a_1,...,a_n$ are integers. If $r$ is a rational zero of $f$, then $r$ is of the form $\pm \frac{p}{q}$, where $p$ is a factor of the constant term $a_0$, and $q$ is a factor of the leading coefficient $a_n$.

---

**Theorem 3.10. Descartes' Rule of Signs:** Suppose $f(x)$ is the formula for a polynomial function written with descending powers of $x$.

- If $P$ denotes the number of *variations of sign* in the formula for $f(x)$, then the number of positive real zeros (counting multiplicity) is one of the numbers $\{ P,P-2,P-4,... \}$.
- If $N$ denotes the number of *variations of sign* in the formula for $f(-x)$, then the number of negative real zeros (counting multiplicity) is one of the numbers $\{ N,N-2,N-4,... \}$.

---

**Theorem 3.11. Upper and Lower Bounds:** Suppose $f$ is a polynomial of degree $n\geq 1$.

- If $c\gt 0$ is synthetically divided into $f$ and all of the numbers in the final line of the division tableau have the same signs, then $c$ is an upper bound for the real zeros of $f$. That is, there are no real zeros greater than $c$.
- If $c\lt 0$ is synthetically divided into $f$ and the numbers in the final line of the division tableau alternate signs, then $c$ is a lower bound for the real zeros of $f$. That is, there are no real zeros less than $c$.

  - **Note:** If the number $0$ occurs in the final line of the division tableau in either of the above cases, it can be treated as $(+)$ or $(-)$ as needed.

### 11\. $f(x)=x^3-2x^2-5x+6$



**Solution.** Possible rational zeros are $\{ \pm 1,\pm 2,\pm 3,\pm 6 \}$.
Try zero 1:
$$
\begin{array}{c}
1 & |
& 1 & -2 & -5 & 6 \\
&& ‚Üì & 1 & -1 & -6 \\ \hline
&& 1 & -1 & -6 & \boxed{0}
\end{array}
$$
therefore $1$ is a zero of $f(x)$, and we get $f(x)=(x-1)(x^2-x-6)$.  Using the quadratic formula to find the roots of $x^2-x-6$ gives $x=-2,3$. Hence $f(x)=(x-1)(x+2)(x-3)$, it follows that the zeros are $x=-2,x=1,x=3$ (each has mult. 1).

### 12\. $f(x)=x^4+2x^3-12x^2-40x-32$

### 13\. $f(x)=x^4-9x^2-4x+12$

### 17\. $f(x)=-17x^3+5x^2+34x-10$

**Solution.** $M=2,[-3,3]$. $R=\{ \pm \frac{1}{17},\pm \frac{2}{17},\pm \frac{5}{17},\pm \frac{10}{17}, \pm 1,\pm 2,\pm 5,\pm 10 \}$. $P=\{2,0\},N=\{1\}$.

```python
import sympy as sp
import numpy as np
import matplotlib.pyplot as plt

x = sp.symbols('x')
f = -17*x**3 + 5*x**2 + 34*x - 10

f_lambda = sp.lambdify(x,f,'numpy')
# x_vals = np.linspace(-3,3,1000)
x_vals = np.linspace(-2,2,1000)
# x_vals = np.linspace(-1,1,1000)
y_vals = f_lambda(x_vals)

plt.plot(x_vals, y_vals, label='$f(x)=-17x^3+5x^2+34x-10$')
plt.axhline(0, color='black',linewidth=0.5)
plt.axvline(0, color='black',linewidth=0.5)
plt.title('Function Plot')
plt.xlabel('x')
plt.ylabel('f(x)')
plt.legend()
plt.show()

zeros_with_multiplicity = sp.roots(sp.Eq(f,0),x)
print(zeros_with_multiplicity)
```
