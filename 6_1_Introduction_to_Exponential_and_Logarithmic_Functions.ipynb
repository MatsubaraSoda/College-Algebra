**Definition 6.1.** A function of the form $f(x)=b^x$ where $b$ is a fixed real numaber, $b\gt 0,\ b\ne 1$ is called a **base b exponential function**.

**Theorem 6.1. Properties of Exponential Functions:** Suppose $f(x)=b^x$.

- The domain of $f$ is $(-\infty,\infty)$ and the range of $f$ is $(0,∞)$.

- $(0,1)$ is on the graph of $f$ and $y=0$ is a horizontal asymptote to the graph of $f$.

- $f$ is one-to-one, continuous and smooth.

- If $b\gt 1$:

  - $f$ is always increasing

  - As $x\to -\infty$, $f(x)\to 0^{+}$

  - As $x\to \infty$, $f(x)\to \infty$

- If $0\lt b\lt 1$:

  - $f$ is always increasing

  - As $x\to -\infty$, $f(x)\to ∞$

  - As $x\to \infty$, $f(x)\to 0^{+}$






```python
import numpy as np
import matplotlib.pyplot as plt

x = np.linspace(-4, 4, 1000)

f_1 = np.power(2, x)
f_2 = np.power(1/2, x)

plt.plot(x, f_1, label=r'$y=2^x$')
plt.plot(x, f_2, label=r'$y=(\frac{1}{2})^x$')

plt.axhline(0, color='black', linewidth=0.25)
plt.axvline(0, color='black', linewidth=0.25)
plt.legend()
plt.show()
```

**Definition 6.2.** The inverse of the exponential function $f(x)=b^x$ is called the **base $\boldsymbol{b}$ logarithm function**, and is denoted $f^{-1}(x)=\log_{b}(x)$ We read '$\log_{b}(x)$' as 'long base $b$ of $x$'.

**Definition 6.3.** The **common logarithm** of a real number $x$ is $\log_{10}(x)$ and is usually written $\log(x)$. The **natural logarithm** of a real number $x$ is $\log_{e}(x)$ and is usually written $\ln(x)$.

**Theorem 6.2. Properties of Logarithmic Functions:** Suppose $f(x)=\log_{b}(x)$.

- The domain of $f$ is $(0,∞)$ and the range of $f$ is $(-∞,∞)$.

- $(1,0)$ is on the graph of $f$ and $x=0$ is a vertical asymptote of the graph of $f$.

- $f$ is one-to-one, continuous and smooth.

- $b^a=c$ if and only if $\log_{b}(c)$ is the exponent you put on $b$ to obtain $c$.

- $\log_{b}(b^x)=x$ for all $x$ and $b^{\log_{b}(x)}=b$ for all $x\gt 0$.

- If $b\gt 1$:

  - $f$ is always increasing

  - As $x\to 0^{+},\ f(x)\to -∞$

  - As $x\to ∞,\ f(x)\to ∞$

- If $0\lt b\lt 1$:

  - $f$ is always decreasing

  - As $x\to 0^{+},\ f(x)\to ∞$

  - As $x\to ∞,\ f(x)\to -∞$


```python
import warnings

x = np.linspace(0, 4, 1000)

with warnings.catch_warnings():
  warnings.simplefilter("ignore", category=RuntimeWarning)
  f_1 = np.log2(x)
  f_1[np.isinf(f_1)] = np.nan

with warnings.catch_warnings():
  warnings.simplefilter("ignore", category=RuntimeWarning)
  f_2 = np.log(x) / np.log(1/2)
  f_2[np.isinf(f_2)] = np.nan

plt.plot(x, f_1, label=r'$y=\log_{2}(x)$')
plt.plot(x, f_2, label=r'$y=\log_{\frac{1}{2}}(x)$')

plt.axhline(0, color='black', linewidth=0.25)
plt.axvline(0, color='black', linewidth=0.25)
plt.legend()
plt.show()
```
