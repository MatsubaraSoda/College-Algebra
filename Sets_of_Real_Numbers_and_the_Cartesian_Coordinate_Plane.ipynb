**Definition 1.1.** A set is a well-defined collection of objects which are called the 'element' of the set. Here, 'well_defined' means that it is possible to determine if something belongs to the collection or not, without prejudice.


---


<center><b>Ways to Describe Sets</b></center>

1. **The Verbal Methond:** Use a sentence to define a set.
2. **The Roster Method:** Begin with a left brace '$\{$', list each element of the set *only once* and then end with a right brace '$\}$'.
3. **The Set-Builder Method:** A combination of the verbal and roster methods using a "dummy variable" such as $x$.

---

<center><b>Sets of Numbers</b></center>

1. The **Empty Set:** $\emptyset = \{\}=\{ x\mid x\ne x \}$. THis is the set with no elements. Like the number '0', it plays a vital role in mathematics.
2. The **Natural Numbers:** $\Bbb{N}=\{1,2,3,...\}$ The periods of ellipsis here indicate that the natural numbers contain 1,2,3, 'and so forth'.
3. The **Whole Numbers:** $\Bbb{W}=\{0,1,2,...\}$
4. The **Integers:** $\Bbb{Z}=\{...,-3,-2,-1,0,1,2,3,...\}$
5. The **Rational Numbers:** $\Bbb{Q}=\{ \frac{a}{b} \mid a\in \Bbb{Z}\ and\ b\in Bbb{Z} \}$. Rational numbers are the ratios of integers (provided the denominator is not zero!) It turns out that another way to describe the rational numbers is:
$$\Bbb{Q}=\{ x\mid x\ \text{possesses a repeating or terminating decimal representation.} \}$$
6. The **Real Numbers:** $\Bbb{R}=\{ x\mid x\ \text{possesses a decimal representation.} \}$
7. The **Irrational Numbers:** $\Bbb{P}=\{ x\mid x\ \text{is a non-rational real number.} \}$ Said another way, an irrational number is a decimal which neither repeats nor terminates.
8. The **Complex Numbers:** $\Bbb{C}=\{ a+bi \mid a,b\in \Bbb{R}\ and\ i=\sqrt{-1} \}$ Despite their importance, the complex numbers play only a minor role in the text.

**Definition 1.2.** Suppose $A$ and $B$ are two sets.

- The **intersection** of $A$ and $B$: $A\cap B=\{ x\mid x\in A \text{ and } x\in B \}$
- The **union** of $A$ and $B$: $A\cup B=\{ x\mid x\in A\text{ or } x\in B \}$

<center><b>Important Facts about the Cartesian Coordinate Plane</b></center>

- $(a,b)$ and $(c,d)$ represent the same point in the plane if and only if $a=c$ and $b=d$.
- $(x,y)$ lies on the $x\text{-axis}$ if and only if $y=0$.
- $(x,y)$ lies on the $y\text{-axis}$ if and only if $x=0$.
- The origin is the point $(0,0)$. It is the only point common to both axes.

---

**Definition 1.3.** Two points $(a,b)$ and $(c,d)$ in the plane are said to be

- **symmetric about the $x\textbf{-axis}$** if $a=c$ and $b=-d$
- **symmetric about the $y\textbf{-axis}$** if $a=-c$ and $b=d$
- **symmetric about the origin** if $a=-c$ and $b=-d$

---

<center><b>Reflections</b></center>

To reflect a point $(x,y)$ about the:

- $x\text{-axis}$, replace $y$ with $-y$.
- $y\text{-axis}$, replace $x$ with $-x$.
- origin, replace $x$ with $-x$ and $y$ with $-y$.

**Equation 1.1. The Distance Formula:** The distance $d$ between the points $P(x_0,y_0)$ and $Q(x_1,y_1)$ is:
$$d=\sqrt{(x_1-x_0)^2+(y_1-y_0)^2}$$

---

**Equation 1.2. The Midpoint Formula:** The midpoint $M$ of the line segment connecting $P(x_0,y_0)$ and $Q(x_1,y_1)$ is:
$$M=(\frac{x_0+x_1}{2},\frac{y_0+y_1}{2})$$

20\. Plot and label the points $A(-3,-7),\ B(1.3,-2),\ C(\pi,\sqrt{10}),\ D(0,8),\ E(-5.5,0),\ F(-8,4),\ G(9.2,-7.8) \text{ and } H(7,5) $ in the Cartesian Coordinate Plane given below.

```python
import matplotlib.pyplot as plt
import numpy as np

# Coordinates of the points
points = {
    'A': (-3, -7),
    'B': (1.3, -2),
    'C': (np.pi, np.sqrt(10)),
    'D': (0, 8),
    'E': (-5.5, 0),
    'F': (-8, 4),
    'G': (9.2, -7.8),
    'H': (7, 5)
}

# Extract x and y coordinates for each point
x_coords, y_coords = zip(*points.values())

# Plot the origin
plt.scatter(0, 0, color='blue', marker='o', s=50, label='Origin')

# Plot the points
plt.scatter(x_coords, y_coords, s=30, color='red')

# Label the points
for point, coords in points.items():
    plt.text(coords[0], coords[1], f'{point} ({coords[0]:.2f},{coords[1]:.2f})', fontsize=8, ha='left')

# Set plot limits
plt.xlim(-10, 10)
plt.ylim(-10, 10)

# Draw the cross-axis grid lines
plt.axhline(0, color='black',linewidth=0.5)
plt.axvline(0, color='black',linewidth=0.5)

# Set axis labels
plt.xlabel('X-axis')
plt.ylabel('Y-axis')

# Add gridlines for better visibility
plt.grid(True, linestyle='--', alpha=0.6)

# Set axis labels
plt.xlabel('X-axis')
plt.ylabel('Y-axis')

# Set plot title
plt.title('Cartesian Coordinate Plane')

# Show the plot
plt.show()
```
