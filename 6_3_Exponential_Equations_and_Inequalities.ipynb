<center><b>Steps for Solving an Equation involving Exponential Functions</b></center>

1. Isolate the exponential function.

2. - If convenient, express both sides with a common base and equate the exponents.
  - Otherwise, take the natural log of both sides of the equation and use the Power Rule.

### $1.$ $2^{4x}= 8$

````python
x = sp.Symbol('x')

lhs = 2**(4*x)
rhs = 8

sol = sp.solve(sp.Eq(lhs, rhs), x)
for s in sol:
  print(s)

x_values = np.linspace(-15, 15, 2000)

lhs_values = sp.lambdify(x, lhs, 'numpy')(x_values)
rhs_values = np.full_like(x_values, 8)

plt.plot(x_values, lhs_values, label=r'lhs=$2^{4x}$')
plt.plot(x_values, rhs_values, label=r'rhs=$8$')

plt.axis('equal')
plt.xlim(-10, 10)
plt.ylim(-10, 10)

plt.axhline(0, color='black', linewidth=0.25)
plt.axvline(0, color='black', linewidth=0.25)
plt.legend()
plt.show()
````
